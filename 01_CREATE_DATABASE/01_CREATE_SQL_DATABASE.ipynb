```python
#import the libs 
import pandas as pd
```

```python
#read every file and save in a single dataframe
def read_raw_data(folder):
    import os 
    file_list=os.listdir(folder)

    raw_data=[]
    for file_name in file_list:
        file_name=f"{folder}/{file_name}"
        file=pd.read_csv(file_name)
        raw_data.append(file)
    raw_data=pd.concat(raw_data)

    return raw_data
    
raw_data=read_raw_data(folder='Sales_Data_RAW')
```

```python
#clean data
def clean_data(df):
    df=df.copy()

    display(df.isna)


    return df
clean_data(raw_data)
```

```python
#separe address column into another ones
def address_cols(df):
    df=df.copy()
    df['Purchase Address']=df['Purchase Address'].map(lambda x:str(x).split(','))
    df['STREET']=df['Purchase Address'].map(lambda x:len(x))

    display(df[df['STREET']!=3])
    return df
address_cols(raw_data)
```
